```python

#check unique values of eval_set
orders['eval_set'].unique()

#droopping the eval_set = test as we do not have target variable(reorder) for this
orders = orders[orders['eval_set'] != 'test']
orders.shape
```

```python
#append order_products_prior and order_products_train
data = order_products_prior.append(order_products_train)
data.shape
```

```python
#left join order_products with orders
data = pd.merge(data, orders, on='order_id', how='left')
data.shape
```

```python
data = pd.merge(data, products, on='product_id', how='left')
print(data.shape)
data.head()
```

```python
#left join data with aisles
data = pd.merge(data, aisles, on='aisle_id', how='left')
print(data.shape)
data.head()
```

```python
#left join data with departments
data = pd.merge(data, departments, on='department_id', how='left')
print(data.shape)
data.head()
```

```python
#moving target column to the end
data = data[[c for c in data if c not in ['reordered']] + ['reordered']]
data.head()
```

```python
#sorting data by user_id and droping the data where user_id > 25 percentile
data2 = data.sort_values('user_id')
data2 = data[data['user_id'] <= data['user_id'].quantile(0.01)]
data2.shape
```

```python
#saving the data as master_data2.csv
data2.to_csv('saved_datasets/master_data2.csv', index=False)
```
